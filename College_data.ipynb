```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
# Read in csv file that contains info about colleges
college_file = "Resources/College.csv"
college_data_df = pd.read_csv(college_file)
college_data_df.head()
```

```python
# Rename college name and grad rate columns in college dataframe
college_data_df = college_data_df.rename(columns={"Unnamed: 0":"college_name", "Grad.Rate":"grad_rate"})
college_data_df                                                              
```

```python
# Read in csv file that contains the city, state and zip codes of colleges
college_location= "Resources/2016 college-scorecard.csv"
college_location_df = pd.read_csv(college_location)
college_location_df.head()
```

```python
# Drop extra columns in  dataframe and keep only applications and graduation rate
new_college_data_df = college_data_df[['college_name', 'Apps', 'grad_rate']].copy()
new_college_data_df.head()
```

```python
# Drop extra columns in dataframe that contains the locations of the colleges
new_college_location_df = college_location_df[['INSTNM', 'CITY', 'STABBR', 'ZIP']].copy()
new_college_location_df.head()
```

```python
# Rename columns in college location dataframe
new_college_location_df = new_college_location_df.rename(columns={"INSTNM":"college_name", "STABBR": "State"})
new_college_location_df                                                              
```

```python
# Drop rows with blanks from the college data dataframe
new_college_data_df.dropna()
```

```python
# Drop rows with blanks from college location dataframe
new_college_location_df.dropna()
```

```python
# Connect to SQL DB
connection_string = "root:Scooter1@localhost/etl_project"
engine = create_engine(f"mysql://{connection_string}")
```

```python
# Confirm tables
engine.table_names()
```

```python
# Write to SQL DB
new_college_data_df.to_sql(name='colleges', con=engine, if_exists='replace', index=False)
```

```python
# Write to SQL DB
new_college_location_df.to_sql(name='address', con=engine, if_exists='replace', index=False)
```

```python
#Confirm data is written to SQL DB
pd.read_sql_query('select * from colleges', con=engine).head()
```

```python
#Confirm data is written to SQL DB
pd.read_sql_query('select * from address', con=engine).head()
```

```python
# Merge two dataframes on institution name
merged_df = pd.merge(new_college_data_df,
                 new_college_location_df,
                 on='college_name', how="left")
merged_df
```
